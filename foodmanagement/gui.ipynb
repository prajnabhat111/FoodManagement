```python
import tkinter as tk               
from tkinter import *
from sklearn.externals import joblib
import pandas as pd


class Frames(object):

    def main_frame(self, root):
        
        tk.Label(root, text="Day: weekday=0 and weekend=1").grid(row=0)
        tk.Label(root, text="Area: 1.Kengeri 2.Jayanagar 3.Kumarswamy Layout 4.Gandhi bazar ").grid(row=1)
        tk.Label(root, text="Time of event: 1.Afternoon 2.Evening 3.Night").grid(row=2)
        tk.Label(root, text ="Type of Event: 1.Birthday party 2.Social event 3.Wedding 4.Pooja").grid(row=3)
        tk.Label(root, text="No of ppl expected:").grid(row=4)

        self.day = tk.Entry(root)
        self.area= tk.Entry(root)
        self.time = tk.Entry(root)
        self.event=tk.Entry(root)
        self.ppl=tk.Entry(root)


        self.day.grid(row=0, column=1)
        self.area.grid(row=1, column=1)
        self.time.grid(row=2, column=1)
        self.event.grid(row=3, column=1)
        self.ppl.grid(row=4, column=1)


        tk.Button(root, text='Quit', command=root.destroy).grid(row=5, column=0, sticky=tk.W, pady=4)
        tk.Button(root, text='Submit',command=self.result_Frame).grid(row=5, column=1, sticky=tk.W, pady=4)

    def result_Frame(self):
        re = Toplevel()
        msg=""
        arr = [0]*13
        
        if(int(self.day.get())==0 or int(self.day.get())==1):
            arr[0]=int(self.day.get())            
        else:
            msg=msg+"  invalid input for day"
    
        if(self.area.get()=="Gandhi bazar"):
            arr[1]=1
        elif(self.area.get()=="Jayanagar"):
            arr[2]=1
        elif(self.area.get()=="Kengeri"):
            arr[3]=1
        elif(self.area.get()=="Kumarswamy Layout"):
            arr[4]=1
        else:
            msg=msg+"  invalid input for area"
   
        if(self.time.get()=="Afternoon"):
            arr[5]=1
        elif(self.time.get()=="Evening"):
            arr[6]=1
        elif(self.time.get()=="Night"):
            arr[7]=1
        else:
            msg=msg+"  invalid input for time"
   
        if(self.event.get()=="Birthday party"):
            arr[8]=1
        elif(self.event.get()=="Pooja"):
            arr[9]=1
        elif(self.event.get()=="Social event"):
            arr[10]=1
        elif(self.event.get()=="Wedding"):
            arr[11]=1  
        else:
            msg=msg+"  invalid input for event"
        
        arr[12]=int(self.ppl.get()) 
    
        if msg == "":
            new=pd.DataFrame([arr],columns=['Day','Gandhi bazar','Jayanagar','Kengeri','Kumarswami layout','Afternoon','Evening','Night','Birthday party','Pooja','Social event','Wedding','No of ppl expected'])
            clf= joblib.load("/Users/prajna/Downloads/foodmanagement/foodmgmtppl.pkl")
            n=clf.predict(new)
            
            rice = joblib.load("/Users/prajna/Downloads/foodmanagement/rice_model.pkl")
            r = rice.predict(n.reshape(-1,1))
            
            sambar = joblib.load("/Users/prajna/Downloads/foodmanagement/sambar_model.pkl")
            s = sambar.predict(n.reshape(-1,1))
            
            roti = joblib.load("/Users/prajna/Downloads/foodmanagement/roti_model.pkl")
            t = roti.predict(n.reshape(-1,1))
            
            sabji = joblib.load("/Users/prajna/Downloads/foodmanagement/sabji_model.pkl")
            b = sabji.predict(n.reshape(-1,1))
            
            payasam = joblib.load("/Users/prajna/Downloads/foodmanagement/payasam_model.pkl")
            p = payasam.predict(n.reshape(-1,1))
            
            tk.Label(re, text="num of ppl who will actually attend :").grid(row=0,column=0)
            tk.Label(re, text=int(n)).grid(row=0,column=1)
            
            tk.Label(re, text="amount of rice required(kgs):").grid(row=1,column=0)
            tk.Label(re, text=int(r)).grid(row=1,column=1)
            
            tk.Label(re, text="amount of sambar required(ltrs) :").grid(row=2,column=0)
            tk.Label(re, text=int(s)).grid(row=2,column=1)
            
            tk.Label(re, text="amount of roti required(in number) :").grid(row=3,column=0)
            tk.Label(re, text=int(t)).grid(row=3,column=1)
            
            tk.Label(re, text="amount of sabji required(kg) :").grid(row=4,column=0)
            tk.Label(re, text=int(b)).grid(row=4,column=1)
            
            tk.Label(re, text="amount of payasam required(ltrs) :").grid(row=5,column=0)
            tk.Label(re, text=int(p)).grid(row=5,column=1)
            
            tk.Button(re, text='Quit', command=re.destroy).grid(row=6, column=3, sticky=tk.W, pady=4) 
        else:
            tk.Button(re, text='Quit', command=re.destroy).grid(row=1, column=3, sticky=tk.W, pady=4)
            tk.Label(re, text=msg).grid(row=0,column=0)

root = tk.Tk()
app = Frames()
app.main_frame(root)
root.mainloop()
```
